### Show distribution of Movies and series with a plot

```python
sns.countplot(data=df, x="type")
```

### Show top15 directors with a barplot

```python
plt.figure(figsize= (15, 7))
top_15 = df["director"].value_counts().drop("Not Given").head(15)
sns.barplot(x=top_15.values, y=top_15.index, linewidth = 1, saturation = 1, color="purple")
plt.title("Top 15 directors", fontsize=14, fontweight="heavy")
plt.show()
```

### Top 15 countries

```python
plt.figure(figsize= (15, 7))
top_15 = df["country"].value_counts().drop("Not Given").head(15)
sns.barplot(x=top_15.values, y=top_15.index, linewidth = 1, saturation = 1, color = "orange")
plt.title("Top 15 countries", fontsize=14, fontweight = "light")
plt.show()
```

### Show all different ratings and its count

```python
plt.figure(figsize = (10,7))
sns.countplot(data=df, x="rating")
plt.title("Rating", fontsize=14, fontweight='light')
plt.show()
```

### Top 40 Genres

```python
plt.figure(figsize= (20, 8))
df_c = df["category"].value_counts().head(40)
sns.barplot(x=df_c.index, y=df_c.values, edgecolor = 'k', linewidth=2, saturation=1, palette='viridis')
plt.xticks(rotation=90)
plt.title("Top 40 Genres", fontweight='light', fontsize=14)
plt.show()
```

### Show release years with boxenplot

```python
plt.figure(figsize= (15, 8))
sns.boxenplot(data=df, y="release_year", color='lightblue')
plt.title("Release Years", fontsize=14, fontweight='light')
plt.show()
```

### Content added through different years

```python
plt.figure(figsize= (20, 8))
df_c = df["release_year"].value_counts().head(15)
sns.lineplot(x=df_c.index, y=df_c.values, color='brown', marker='s')
plt.title("Content added through different years", fontweight="light", fontsize=14)
plt.show()
```
